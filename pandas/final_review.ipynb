```python
#import dependencies
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
csv_file = "tv_shows.csv"
tv_shows_df = pd.read_csv(csv_file)
tv_shows_df.head()
```

```python
show_df = tv_shows_df[['Title', 'Year', 'IMDb', 'Rotten Tomatoes', 'Netflix', 'Hulu', 'Prime Video', 'Disney+']]
```

```python
show_df.head()
```

```python
# Change TV show providers to boolean values
show_df[['Netflix','Hulu', 'Prime Video', 'Disney+']] = show_df[['Netflix','Hulu', 'Prime Video', 'Disney+']].astype(bool)
```

```python
show_df.head()
```

```python
show_df['Rotten Tomatoes'].isnull().sum().sum()
```

```python
show_df['IMDb'].isnull().sum().sum()
```

```python
#Replace missing values in Rotten Tomatoes column
show_df['Rotten Tomatoes'].fillna('00%', inplace=True)
```

```python
show_df['Rotten Tomatoes'].isnull().sum().sum()
```

```python
#Replace missing values in IMDb column
show_df['IMDb'].fillna(0.0, inplace=True)
```

```python
#Replace missing values in 
show_df['IMDb'].isnull().sum().sum()
```

```python
rate_df = show_df
rate_df.head()
```
